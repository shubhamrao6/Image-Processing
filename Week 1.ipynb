```python
def resolution(img, factor):
    n = factor
    resolved_img = np.zeros((int(img.shape[0]/n), int(img.shape[1]/n)))
    for i in range(0, img.shape[0] -1):
        for j in range(0, img.shape[1] - 1):
            resolved_img[int(i/n), int(j/n)] = np.sum(img[i:i+n, j:j+n])
    return resolved_img
```
